```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
```

```python
data
```

# Preprocessing

```python
mappings = list()

encoder = LabelEncoder()
for column in range(len(data.columns)):
    data[data.columns[column]] = encoder.fit_transform(data[data.columns[column]])
    mappings_dict = {index: label for index, label in enumerate(encoder.classes_)}
    mappings.append(mappings_dict)
```

```python
mappings
```

```python
y = data['class']
X = data.drop('class', axis = 1)
```

Set the y (the answer for each row) and removed column 'class' from x (because we do not need to scale the prediction)

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

# Model Selection

```python
nn_model = MLPClassifier(hidden_layer_sizes=(128, 128)) #set 2 hidden layers 128 nodes each
```

# Training

```python
np.sum(y) / len(y) #check whether the dataset is balanced
```

```python
nn_model.fit(X_train,y_train)
```

```python
print(f"------Neural Network: {nn_model.score(X_test, y_test)}")
```
